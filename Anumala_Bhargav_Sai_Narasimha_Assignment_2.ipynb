# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
# Your code here
# (2)
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup

#url for imdb
base_url = "https://www.imdb.com/title/tt15398776/reviews/_ajax"
target_review_count = 1000
# array to store reviews
all_reviews = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'
}

while len(all_reviews) < target_review_count:
    payload = {
        "sort": "curated",
        "dir": "desc",
        "ratingFilter": 0,
        "paginationKey": ''
    }

    # Sending request for ajax url
    response = requests.post(base_url, headers=headers, data=payload)
    print("Response Status Code:", response.status_code)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        review_elements = soup.select('.lister-item-content')

        for review in review_elements:
            text = review.select_one('.text.show-more__control').get_text(strip=True) if review.select_one('.text.show-more__control') else 'No Text'  # Review text
            all_reviews.append({"text": text})

        if len(all_reviews) >= target_review_count:
            break  # break after target
        time.sleep(2)

    else:
        print("Error fetching reviews.")
        break

all_reviews = all_reviews[:target_review_count]
df = pd.DataFrame(all_reviews)
df.to_excel("oppenheimer_1000_reviews.xlsx", index=False, header=True)

print(f"Successfully scraped {len(all_reviews)} reviews and saved to 'oppenheimer_1000_reviews.xlsx'.")
```

```python
# (6)
import pandas as pd
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')  #wordNet
from nltk.corpus import stopwords

data_frame = pd.read_excel('oppenheimer_1000_reviews.xlsx', header=0)
data_frame.rename(columns={'text': 'review'}, inplace=True)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def sanitize_review(input_review):
    input_review = input_review.lower()
    no_numbers = re.sub(r'\d+', '', input_review)
    sanitized = re.sub(r'[^\w\s]', '', no_numbers)
    sanitized = re.sub(r'\s+', ' ', sanitized).strip()
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in sanitized.split() if word not in stop_words]
    stemmed_words = [stemmer.stem(word) for word in filtered_words]
    lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words] #lemmatization
    return ' '.join(lemmatized_words)

data_frame['cleaned_review'] = data_frame['review'].apply(sanitize_review)
data_frame.to_csv('cleaned_reviews_full_processing.csv', index=False)  # Saving as CSV without the index
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Your code here
import pandas as pd
import nltk
import spacy
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
data_frame = pd.read_csv('cleaned_reviews_full_processing.csv')
nlp = spacy.load("en_core_web_sm")

# pos tagging
def pos_tagging(text):
    tokens = nltk.word_tokenize(text)
    pos_tags = nltk.pos_tag(tokens)
    pos_count = {'N': 0, 'V': 0, 'Adj': 0, 'Adv': 0}

    for word, tag in pos_tags:
        if tag.startswith('N'):
            pos_count['N'] += 1
        elif tag.startswith('V'):
            pos_count['V'] += 1
        elif tag.startswith('J'):
            pos_count['Adj'] += 1
        elif tag.startswith('R'):
            pos_count['Adv'] += 1

    return pos_tags, pos_count

#parsing
def parse_sentence(sentence):
    doc = nlp(sentence)
    return doc

def extract_entities(text):
    doc = nlp(text)
    entities = {
        'PERSON': 0,
        'ORG': 0,
        'GPE': 0,
        'PRODUCT': 0,
        'DATE': 0
    }

    for ent in doc.ents:
        if ent.label_ in entities:
            entities[ent.label_] += 1

    return entities

for index, row in data_frame.iterrows():
    review = row['cleaned_review']
    pos_tags, pos_count = pos_tagging(review)
    print(f"Review {index + 1} POS Tags: {pos_tags}")
    print(f"POS Count: {pos_count}\n")
    sentences = nltk.sent_tokenize(review)
    for sentence in sentences:
        constituency_tree = nltk.Tree.fromstring(str(nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sentence)))))
        doc = parse_sentence(sentence)
        print(f"Dependency Parse Tree for '{sentence}':")
        for token in doc:
            print(f"{token.text} -> {token.dep_} ({token.head.text})")

        print("\nConstituency Parse Tree:")
        print(constituency_tree)
        print("\n")

    entities = extract_entities(review)
    print(f"Entity Counts in Review {index + 1}: {entities}\n")

print("Completed")
```
