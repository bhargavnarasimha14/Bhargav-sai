## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.sentiment import SentimentIntensityAnalyzer
from gensim.models import Word2Vec
import nltk
# downloading vader from nltk
nltk.download('vader_lexicon')
# Loading the dataset
data = pd.read_csv('IMDB Dataset.csv')
# text preprocessing
def preprocesstext(text):
    text = re.sub(r'<br />', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower().strip()
    return text
# review preprocessing
data['cleaned_review'] = data['review'].apply(preprocesstext)
print("\nPreprocessed Reviews:")
print(data['cleaned_review'].head())
# Bag of Words
bow_vectorizer = CountVectorizer()
bow_features = bow_vectorizer.fit_transform(data['cleaned_review'])
print("\nBag of Words Features:")
print(bow_features.toarray())
print("Feature Names (BoW):", bow_vectorizer.get_feature_names_out())
# termfrequency-inverse document frequenct
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(data['cleaned_review'])
print("\nTF-IDF Features:")
print(tfidf_features.toarray())
print("Feature Names (TF-IDF):", tfidf_vectorizer.get_feature_names_out())
sia = SentimentIntensityAnalyzer()
data['sentiment_score'] = data['cleaned_review'].apply(lambda x: sia.polarity_scores(x)['compound'])
print("\nSentiment Polarity Scores:")
print(data[['cleaned_review', 'sentiment_score']].head())
tokenized_reviews = [review.split() for review in data['cleaned_review']]
word2vec_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=1, workers=4)
# output
print("\nWord Embeddings of the word ""film"":")
print(word2vec_model.wv['film'])
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
# creating sentiment column
data['sentiment'] = data['sentiment_score'].apply(lambda x: 'positive' if x > 0 else 'negative')
# termfrequency-inverse document frequency
# adding max feautures
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X = tfidf_vectorizer.fit_transform(data['cleaned_review'])
le = LabelEncoder()
y = le.fit_transform(data['sentiment'])
chi2_scores, p_values = chi2(X, y)
# Creating data frame
feature_scores = pd.DataFrame({
    'feature': tfidf_vectorizer.get_feature_names_out(),
    'chi2_score': chi2_scores,
    'p_value': p_values
})
feature_scores = feature_scores.sort_values('chi2_score', ascending=False)
print("Top 20 most important features:")
print(feature_scores.head(20))
n_features_to_keep = 1000
top_features = feature_scores.head(n_features_to_keep)['feature'].tolist()
# creating tf-idf matrix
tfidf_vectorizer_selected = TfidfVectorizer(vocabulary=top_features)
X_selected = tfidf_vectorizer_selected.fit_transform(data['cleaned_review'])
print(f"\nShape of the feature matrix after selection: {X_selected.shape}")
```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertModel
import torch
# loading the dataset
data = pd.read_csv('IMDB Dataset.csv')
# getting bert embeddings
def get_bert_embedding(text, model, tokenizer):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
# loading bert
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
print("Generating BERT embeddings")
embeddings = []
for review in data['review'][:1000]:  # Limit to first 1000 reviews for efficiency
    embedding = get_bert_embedding(review, model, tokenizer)
    embeddings.append(embedding)
# convert to array
embeddings = np.array(embeddings)
query = "A great science fiction movie with amazing special effects"
query_embedding = get_bert_embedding(query, model, tokenizer)
similarities = cosine_similarity([query_embedding], embeddings)[0]
data['similarity'] = similarities
data_sorted = data.sort_values('similarity', ascending=False)
print("\nTop 10 most similar reviews to the given query:")
for i, row in data_sorted.head(10).iterrows():
    print(f"Similarity: {row['similarity']:.4f}")
    print(f"Review: {row['review'][:200]}...")
    print()
data_sorted.to_csv('similarityresults.csv', index=False)
print("results saved as a csv file")
```
